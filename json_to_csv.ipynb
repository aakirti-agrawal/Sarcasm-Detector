```python
import numpy as np
import pandas as pd
import json
import os
os.listdir()
```

```python
json_obj = []
for line in open('train.json', 'r'):
    json_obj.append(json.loads(line))
json_obj
```

```python
print(len(json_obj))
print(json_obj[0].keys())
json_obj[0]
```

```python
df = pd.DataFrame.from_dict(json_obj, orient='columns')
print(df.shape)
df.head(10)
```

```python
df.to_csv('train.csv')
```
